```python
import pandas as pd

df1 = pd.read_csv('part-00000', header=None)
```

```python
df1.head()
```

```python
for i in range(len(df1)):
    df1.loc[i][0] = df1.loc[i][0].replace('(', '')
    df1.loc[i][1] = df1.loc[i][1].replace(')','').replace(' ','')
    df1.loc[i][2] = df1.loc[i][2].replace(')','').replace(' ','')
```

```python
df1 = df1.sort_values(by=0)
df1.head()
```

```python
start = 0
end = 6

while end <= 120000:
    df2['Predictor'][df2[start:end]['Rating'].sort_values(ascending=False).index[0:3]] = 1
    
    start = end
    end = end+6
```
